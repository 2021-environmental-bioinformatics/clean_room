```r
metadata<-read.table(file = 'metadata', sep = '\t', header = TRUE) #read in metadata file
View(metadata)
sampledata = sample_data(metadata) #assign metadata file as sample_data in phyloseq class object (sample variables)
```

```r
sample_names(sampledata)<-c("2_10", "2_2", "2_4","2_5","2_6","2_7","2_8","2_9","3_1","3_3","3_4","3_5","3_6","3_8","3_9","4_1","4_2","4_3","4_4","4_6","4_7","4_8")
#set sample names to match phyloseq object for merging (can't figure out how to do this automatically)
```

```r
merged = merge_phyloseq(merged_metagenomes, sampledata) #merge phyloseq object with metadata
View(merged) #check that it worked
sample_data(merged)$merged <- get_variable(merged, "Grouped") #set grouping variable for later use
merged_mean = merge_samples(merged, "Grouped") #combine samples in same group to get mean
```

```r
View(merged@tax_table@.Data) #view taxonomic data
```

```r
merged@tax_table@.Data <- substring(merged_metagenomes@tax_table@.Data, 4) #reassign column names for tax table to taxonomy levels
colnames(merged@tax_table@.Data)<- c("Kingdom", "Phylum", "Class", "Order", "Family", "Genus", "Species")
merged = subset_taxa(merged, Phylum!="Chordata")
```

```r
unique(merged@tax_table@.Data[,"Phylum"]) #look at all the unique phyla represented in the data
```

```r
options(repr.plot.width=15, repr.plot.height=8)
plot_richness(physeq = merged, x="Grouped", measures = c("Observed","Shannon"), color="Grouped")+
    geom_boxplot(alpha=0.6)+ 
    theme(legend.position="none", axis.text.x=element_text(angle=45,hjust=1,vjust=1,size=12)) +theme_bw()#plot species richness and shannon entropy in box plots, using grouping variable for X-axis. Use default sections.
```

```r
deep <- data.frame(Samples = sample_names(merged_metagenomes), Reads = sample_sums(merged_metagenomes)) #convert to dataframe and check reads depth for each of the samples 
```

```r
ggplot(data = deep, mapping = aes(x = Samples,y = Reads)) +
  geom_col() #plot reads depth from dataframe as bar plot
```

```r
top_merged=get_top_taxa(merged, 12, relative = TRUE, discard_other = FALSE,
  other_label = "Other")
top_merged=name_taxa(top_merged, other_label = NULL,
  species = TRUE, unique_rank = NULL, unique_sep = " ")
```

```r
percentages  = transform_sample_counts(merged, function(x) x*100 / sum(x) ) #conver sample counts for each taxa to percentages
head(percentages@otu_table@.Data) #view head of percentages table (otu part of phyloseq object)
```

```r
percentagestop  = transform_sample_counts(top_merged, function(x) x*100 / sum(x) ) #conver sample counts for each taxa to percentages
head(percentages@otu_table@.Data) #view head of percentages table (otu part of phyloseq object)
```

```r
meta.ord <- ordinate(physeq = percentages, method = "PCoA", 
                     distance = "bray")#prepare PCA data using bray distance metric from percentages data. 
```

```r
plot_ordination(physeq = percentages, color="Grouped", ordination = meta.ord)#plot ordination data using 'Grouped' sample varaible to assign colour
```

```r
glom <- tax_glom(percentagestop, taxrank = 'Species') # group reads from a set taxonomic rank together, here, all samples in the same phylum
View(glom@tax_table@.Data) #view head of cut table - should only have phylum remaining. 
```

```r
percentages <- psmelt(glom) #melt phyloseq object into large dataframe for use in ggplot3
str(percentages) #display structure of object
```

```r
raw <- tax_glom(physeq = top_merged, taxrank = "Species") #take raw data from glommed table 
raw.data <- psmelt(raw) #convert to dataframe
str(raw.data) #check structure
```

```r
raw.plot <- ggplot(data=raw.data, aes(x=Sample, y=Abundance, fill=Species))+ 
  geom_bar(aes(), stat="identity", position="stack",color='black')+theme_bw() #using sample as x and abundance as y, create stacked bar plot for absolute number of reads per phylum per sample
```

```r
library(umap)
s<- get.samp(merged)
otu <- get.otu(merged,as.matrix=TRUE) %>% t()
umap <- umap(otu)
umapdata <- umap$layout %>% data.frame() %>% rownames_to_column("sample") %>% 
  rename(UMAP1=X1,UMAP2=X2) %>% left_join(s,by="sample")
ggplot(umapdata,aes(x=UMAP1,y=UMAP2,color=Consistency)) + 
  geom_point() +
  theme(aspect.ratio=1)
```

```r
mergedones  = transform_sample_counts(merged, function(x) x / (x+0.0000000000000001) ) #conver sample counts for each taxa to binary presence/absence
View(mergedones@otu_table@.Data)
```

```r
meta.ord <- ordinate(physeq = mergedones, method = "CCA", 
                     distance = "bray")
```

```r
plot_ordination(physeq = mergedones, color="Grouped", ordination = meta.ord)+geom_point(size=8)+theme_bw()#plot ordination data using 'Grouped' sample varaible to assign colour
```

```r
tsne_obj<-tsne_phyloseq(mergedones, distance = "bray", perplexity = 10
                        , dimensions = 2,
  precomputed_distance = NULL, pseudocounts = 1, verbose = 1,
  rng_seed = NULL, philr_options = list(), control = list())
```

```r
plot_tsne_phyloseq(mergedones, tsne_obj, axes = 1:2, color = "Grouped",
  shape = NULL, title = NULL, justDF = FALSE)+geom_points(size=8)
```
